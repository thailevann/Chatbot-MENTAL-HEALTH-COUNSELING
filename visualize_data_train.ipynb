```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datasets import load_dataset

# Tải dữ liệu từ Hugging Face
dataset = load_dataset("thailevann/mental_health_vi_1")

# Chuyển đổi thành DataFrame
df = pd.DataFrame(dataset['train'])

# Hiển thị 5 dòng đầu tiên của dữ liệu
print(df.head())

# Kiểm tra thông tin cơ bản của dữ liệu
print(df.info())

# Hiển thị phân phối độ dài của cột Context và Response
df['Context_length'] = df['Context_translated'].apply(len)
df['Response_length'] = df['Response_translated'].apply(len)

# Biểu đồ phân phối độ dài của Context và Response
plt.figure(figsize=(14, 7))

plt.subplot(1, 2, 1)
sns.histplot(df['Context_length'], kde=True)
plt.title('Phân phối độ dài của Context')

plt.subplot(1, 2, 2)
sns.histplot(df['Response_length'], kde=True)
plt.title('Phân phối độ dài của Response')

plt.show()

# Hiển thị một vài cặp Context và Response
for i in range(5):
    print(f"Context {i+1}: {df['Context_translated'].iloc[i]}")
    print(f"Response {i+1}: {df['Response_translated'].iloc[i]}")
    print("-" * 80)
```

```python
max_context_length = df['Context_length'].max()
min_context_length = df['Context_length'].min()
max_response_length = df['Response_length'].max()
min_response_length = df['Response_length'].min()
```

```python
print(f"Độ dài lớn nhất của Context: {max_context_length}")
print(f"Độ dài nhỏ nhất của Context: {min_context_length}")
print(f"Độ dài lớn nhất của Response: {max_response_length}")
print(f"Độ dài nhỏ nhất của Response: {min_response_length}")
```

```python
import pandas as pd
from datasets import load_dataset
from collections import Counter
import matplotlib.pyplot as plt
import seaborn as sns

# Tải dữ liệu từ Hugging Face
dataset = load_dataset("thailevann/mental_health_vi_1")

# Chuyển đổi thành DataFrame
df = pd.DataFrame(dataset['train'])

# Kết hợp tất cả các văn bản trong cột Context
all_context_text = ' '.join(df['Context_translated'].tolist())

# Tách các từ và đếm tần suất từ trong cột Context
context_words = all_context_text.split()
context_word_counts = Counter(context_words)

# Lấy top 20 từ xuất hiện nhiều nhất trong cột Context
top_20_context_words = context_word_counts.most_common(20)
top_20_context_df = pd.DataFrame(top_20_context_words, columns=['Từ', 'Số lượng'])

# Kết hợp tất cả các văn bản trong cột Response
all_response_text = ' '.join(df['Response_translated'].tolist())

# Tách các từ và đếm tần suất từ trong cột Response
response_words = all_response_text.split()
response_word_counts = Counter(response_words)

# Lấy top 20 từ xuất hiện nhiều nhất trong cột Response
top_20_response_words = response_word_counts.most_common(20)
top_20_response_df = pd.DataFrame(top_20_response_words, columns=['Từ', 'Số lượng'])
```

```python
# Thiết lập kiểu cho biểu đồ
sns.set(style="whitegrid")

# Tạo các trục con cho hai biểu đồ
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(12, 16))

# Biểu đồ cho cột Context
sns.barplot(ax=axes[0], x='Số lượng', y='Từ', data=top_20_context_df, palette='viridis')
axes[0].set_title('Top 20 từ xuất hiện nhiều nhất trong cột Context', fontsize=16)
axes[0].set_xlabel('Số lượng', fontsize=14)
axes[0].set_ylabel('Từ', fontsize=14)

# Biểu đồ cho cột Response
sns.barplot(ax=axes[1], x='Số lượng', y='Từ', data=top_20_response_df, palette='viridis')
axes[1].set_title('Top 20 từ xuất hiện nhiều nhất trong cột Response', fontsize=16)
axes[1].set_xlabel('Số lượng', fontsize=14)
axes[1].set_ylabel('Từ', fontsize=14)

# Tăng khoảng cách giữa các trục con
plt.tight_layout()

# Hiển thị biểu đồ
plt.show()
```
